## Abstract

In Statistics, to create meaningful model, a tidy or clean data set is extremely important. So, data examination and cleaing is the pre-face of any modelling creation. Data cleaning is the process of detecting and correcting (or removing) corrupt or inaccurate records from a record set, table, or database and refers to identifying incomplete, incorrect, inaccurate or irrelevant parts of the data and then replacing, modifying, or deleting the dirty or coarse data. In this unit, we would experiment with data imputation process of replacing missing data with substituted value using the Boston Housing Dataset. The data consists of price of houses in various places in Boston. We would then apply different types of missingness, such as Missing Completely at Random (MACR), Missing at Random (MAR) and Missing Not at Random (MNAR) to compare the (linear) model performance in predicting Boston Housing prices.

```python
%matplotlib inline 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn


import seaborn as sns
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
bos = pd.DataFrame(boston.data)

bos.columns = boston.feature_names
bos.head()
```

```python
df1 = bos[bos.isnull().any(axis=1)]
df1
```

```python
bos['PRICE'] = boston.target
bos.head()
```

```python
bos.describe()
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
X = bos.iloc[:,0:-1]  #iloc gives the positional locations, so 'all rows' and 'all but the last column'
Y_true = bos.iloc[:,-1] 
lm.fit(X,Y_true)

Y_pred = lm.predict(X)

plt.scatter(Y_true, Y_pred)
plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Prices vs Predicted prices: $Y_i$ vs $\hat{Y}_i$")
```

```python
lm.score(X,Y_true)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print(mean_squared_error(Y_pred,Y_true))
print(mean_absolute_error(Y_pred,Y_true))
#lm.score(X,Y_true)
```

## Step 3 - Missing at Random --? MAR between 0-10%

```python
X_twocol = X
X2 = X_twocol[X_twocol['CHAS']==1]
```

```python
df1 = X[X.isnull().any(axis=1)]
df1
```

```python
limit = .10
for i in range(X2.shape[0]):
    miss_1 = random.random()
    miss_2 = random.random()
    print(miss_1)
    print(miss_2)
    if miss_1 < limit:
        X2.iloc[i,4] = np.nan
        
    if miss_2 < limit:
        X2.iloc[i,7] = np.nan
```

```python
X2
```

```python
sub = X_twocol['NOX'].mean()
```

```python
X_twocol.loc[X_twocol.iloc[:,4].isna(),'NOX'] = sub
```

```python
sub = X_twocol['DIS'].mean()

X_twocol.loc[X_twocol.iloc[:,7].isna(),'DIS'] = sub
```

## Step 4 - Missing at not Random

```python
#5.885500 - Missing at Not Random
X_rm = X
#X2 = X_rm[X_twocol['CHAS']==1]
```

```python
len(X[X['RM']<5.8855])
```

```python
X2 = X_rm[X_rm['RM']<=6.208]
```

```python
limit = .25
for i in range(X2.shape[0]):
    miss_1 = random.random()
    print(miss_1)
    if miss_1 < limit:
        X2.iloc[i,5] = np.nan
```

```python
X_rm[X_rm['RM']<=6.208] = X2
```

```python
sub =  X_rm['RM'].mean()
X_rm.loc[X_rm.iloc[:,5].isna(),'RM'] = sub
```

```python
from sklearn.linear_model import LinearRegression

lm.fit(X_rm,Y_true)

Y_pred = lm.predict(X_rm)

plt.scatter(Y_true, Y_pred)
plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Prices vs Predicted prices: $Y_i$ vs $\hat{Y}_i$")
```

```python
lm.score(X_rm,Y_true)
```

```python
print(mean_squared_error(Y_pred,Y_true))
print(mean_absolute_error(Y_pred,Y_true))
```
